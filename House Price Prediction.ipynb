```python
x = df.drop(columns = ['price'])
y = df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

## applying Linear Regression


```python
column_trans = make_column_transformer((OneHotEncoder(sparse_output=False), ['beds']), remainder='passthrough')
```

```python
scaler = StandardScaler()
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
from sklearn.preprocessing import StandardScaler  # ← Correct capitalization
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)

from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_scaled, y)
```

```python
pipe = make_pipeline(column_trans, scaler, lr)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_lr = pipe.predict(x_test)
```

```python
r2_score(y_test, y_pred_lr)
```

## using lasso 
### (Least Absolute Shrinkage and Selection Operator):It minimizes the usual sum of squared errors, plus the absolute values of coefficients multiplied by a penalty term alpha.

```python
lasso = Lasso()
```

```python
pipe = make_pipeline(column_trans, scaler, lasso)
pipe.fit(x_train, y_train)
```

```python
y_pred_lasso = pipe.predict(x_test)
r2_score(y_test, y_pred_lasso)
```

## using ridge
### Ridge Regression is a type of linear regression that adds L2 regularization to the model. This means it penalizes large coefficients in the model to reduce overfitting and improve generalization.

```python
ridge = Ridge()
```

```python
pipe = make_pipeline(column_trans, scaler, ridge)
```

```python
pipe.fit(x_train, y_train)
```

```python
y_pred_ridge = pipe.predict(x_test)
r2_score(y_test, y_pred_ridge)
```

```python
# Comparing with other models
print("No Regularization: ", r2_score(y_test, y_pred_lr))
print("Lasso: ", r2_score(y_test, y_pred_lasso))
print("Ridge: ", r2_score(y_test, y_pred_ridge))
```
